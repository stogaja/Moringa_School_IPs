### IMPORTING NUMPY AND PANDAS

```python
# Let's import the pandas library and Numpy library
import pandas as pd
import numpy as np
```

##### DATA CLEANING TO CORRECT SINGULAR COLUMN IN CELL_GEO DATASET

```python
# Let's read the data from the csv file and add a separator
data = pd.read_csv('cells_geo.csv', sep=';')
data.head()
```

```python
# let's rename the SITE_CODE column to SITE_ID
data.rename(columns={'SITE_CODE': 'SITE_ID'}, inplace=True)
data.head()
```

```python
# let's read the data from the csv file
df1 = pd.read_csv('Telcom_dataset.csv')
df1.head()
```

```python
# let's read the data from the csv file
df2 = pd.read_csv('Telcom_dataset2.csv')
df2.head()
```

```python
# let's read the data from the csv file
df3 = pd.read_csv('Telcom_dataset3.csv')
df3.head()
```

```python
# Rename the Days dataset columns to be common
new_columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_NUMBER_INT', 
           'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID']

df1.columns = new_columns
df2.columns = new_columns
df3.columns = new_columns
```

```python
# let's merge the datasets
df = pd.concat([df1, df2, df3])
df.head()
```

```python
# let's merge data and df together
df = pd.merge(data, df, on='SITE_ID')
df.head()
```

#### Renaming the Villes to city

```python
# Let's rename the Villes column to City
data.rename(columns={'VILLES': 'City'}, inplace=True)
data.head()
```
