### IMPORTING NUMPY AND PANDAS

```python
# Let's import the pandas library and Numpy library
import pandas as pd
import numpy as np
```

##### DATA CLEANING TO CORRECT SINGULAR COLUMN IN CELL_GEO DATASET

```python
# Let's read the data from the csv file and add a separator
data = pd.read_csv('cells_geo.csv', sep=';')
data.head()
```

#### REMOVING SEMI COLONS FROM COLUMN NAMES

```python
# Let's remove the semicolon from the column names and rows
data.columns = data.columns.str.replace(';', '')
data.head()
```

#### Renaming the Villes to city

```python
# Let's rename the Villes column to City
data.rename(columns={'VILLES': 'City'}, inplace=True)
data.head()
```
