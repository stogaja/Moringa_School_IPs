```python
# QUESTION 1: FINDING THE CHANGES IN MORTALITY RATE FROM WHEN THERE WERE NO VACCINE TO WHEN THERE WAS A VACCINE

# let's find the mortality when there were no vaccines at the time of the data
var1 = data[data['total_vaccinations'] == 0]

var1.groupby(['continent'])['total_deaths'].sum().sort_values(ascending = False)
```

```python
# let's find the mortality when there were an abundance of vaccines and add the country names

var2 = data[data['total_vaccinations'] > 0]

var2.groupby(['continent'])['total_deaths'].sum().sort_values(ascending = False)
```

#### The inference for the investigation above has been documented below

```python
# In the data provided, certain regions such as Afghanistan had no data provided on mortality rates at the start
# so our inference is that the mortality rate was 0 then. We can see this in the results above. We can also see that 
# the mortality rised to a maximum of 5886957.0 in the data provided and then fell again when there was an abundance of vaccines.
```

```python
# QUESTION 2: FINDING THE AREAS WITH LOW VACCINE COVERAGE

# let's find the areas with low vaccine supply and print the first 10 rows

data['total_vaccinations'].groupby(data['location']).min().sort_values(axis=0, ascending=False).head(10)
```

```python
# QUESTION 4:

# Let’s find the overall change in population from the COVID-19 pandemic in:
# Kenya, Malaysia, Spain, Venezuela, Panama

data['population_diff'] = data['population'] - data['total_deaths']
data.tail()

country = data[(data['location'] == 'Kenya')]
change_kenya = (country['population_diff'].max() - country['population_diff'].min())
overall_change = (change_kenya / country['population_diff'].max() *100)

print("The overall change in population was", change_kenya, "and it by", overall_change, "percent")
```

```python
# QUESTION 5:
# Let's investigate whether diabetes prevalence have adverse effects prevalent amongst COVID-19 victims post the pandemic in Kenya, Malaysia, Spain, Venezuela, Panama. 
#
column_1 = data["total_cases"]

column_2 = data["diabetes_prevalence"]

correlation = column_1. corr(column_2)

print(correlation)
```

```python
#Let’s investigate whether cardio-vascular diseases have adverse effects amongst COVID-19 victims post the pandemic in Kenya, Malaysia, Spain, Venezuela, Panama. 
#
column_1 = data["total_deaths"]

column_2 = data["cardiovascular_death_rate"]

correlation = column_1. corr(column_2) 

print(correlation)
```

#### The inference for the correlation analysis has been documented below.

```python
# It is evident that diabetes prevalence had a positive effect on the total number of COVID-19 cases. This implies that the diabetes prevalence increased the likelihood of one contacting COVID-19.
# It is also evident that cardio-vascular death rate had a negative effect on the total number of COVID-19 deaths. This implies that cardio-vascular death rate decreased the likelihood of one dying of COVID-19.
```
